#### Menampilan Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./dataset/housing.csv")

df.head(10)
```

#### Visualisasi Data-1: Data Besaran Penduduk Berdasarkan lokasinya
Pada data dibawah ini menunjukkan besaran penduduk dilihat dari lokasinya, terlihat bahwa sebagian besar tempat mempunyai populasi lebih dari 5000 penduduk, sedangkan hanya sedikit tempat yang mempunyai populasi lebih dari 15000 penduduk.

```python
plt.figure(figsize=(15,10))
plt.scatter(df['latitude'],df['longitude'],c=df['population'], cmap='Spectral')
plt.colorbar().set_label("Population")
plt.title('Population Magnitude in Different Areas')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```

#### Visualisasi Data-2: Data Jumlah Rumah Berdasarkan Usia Rumah

Dibawah ini kita bisa melihat Jumlah rumah berdasarkan Rumah. Terlihat bahwa usia maksimal rumah yang ada di california adalah 52 tahun.


```python
plt.figure(figsize=(18,10))
df['housing_median_age'].value_counts().plot(kind='bar',color='red')
plt.xlabel("Age")
plt.ylabel("Number of Houses")
plt.title("Median Age of Housing")
```

#### Visualisasi Data-3: Data Jumlah Rumah Berdasarkan Lokasinya

Dibawah ini kita bisa melihat Jumlah rumah berdasarkan jarak ke laut. Terlihat bahwa jumlah rumah terbanyak berada 1 jam dari laut, sedangkan jumlah rumah terdikit terdapat di pulau.


```python
plt.figure(figsize=(10,10))
df['ocean_proximity'].value_counts().plot(kind='pie',autopct='%1.2f%%')
plt.title('Preferred Proximity from Ocean')
```
