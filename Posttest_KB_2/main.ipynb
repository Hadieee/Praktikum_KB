#### Menampilan Dataset

```python
import pandas as pd
import numpy as np
from prettytable import PrettyTable, MARKDOWN
from prettytable.colortable import ColorTable, Themes

df = pd.read_csv("./dataset/housing.csv")

df.head(10)
```

##### Ketentuan Analisis Dataset-1: Informasi yang didapatkan adalah jumlah record, nama attribute, jumlah attribute dengan tipe objek, jumlah attribute dengan tipe angka, jumlah attribute yang memiliki nilai null.

*total_records*: Dengan menggunakan fungsi `df.shape`, kita dapat mengembalikan total baris dan kolom.

*attributes*: Dengan menggunakan fungsi`df.columns.values`, kita dapat mengembalikan bentuk numpy array dari fungsi `df.columns` 

*object_attributes*: Dengan menggunakan fungsi `select_dtypes`, kita dapat mengambil baris dengan kolom yang memiliki data type objek, kemudian diambil kolomnya saja

*number_attributes*: Dengan menggunakan fungsi `select_dtypes`, kita dapat mengambil baris dengan kolom yang memiliki data type angka, kemudian diambil kolomnya saja

*null_attribtues*: Dengan fungsi `df.isnull().any()`, kita dapat mengembailkan list kolom dan value boolean tentang apakah ada data yang kosong atau tidak, kita akan mengambil yang bernilai `True` dan mengambil nama kolomnya

```python
info = {
  'total_records': 0,
  'attributes': [],
  'object_attributes': 0,
  'number_attributes': 0,
  'null_attributes': 0
}

info['total_records'] = df.shape[0]
info['attributes'] = list(df.columns.values)
info['null_attributes'] = list(df.loc[:, df.isnull().any()].columns)
info['object_attributes'] = list(df.select_dtypes(['object']).columns)
info['number_attributes'] = list(df.select_dtypes(['float64', 'int64']).columns)
```

#### Ketentuan Analisis Dataset-2: Masukkan informasi di atas dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin

```python
table = PrettyTable(["Data", "Hasil"])
table.align = "c"
table.set_style(MARKDOWN)
table = ColorTable(theme=Themes.OCEAN)

table.add_row(["Jumlah Record", info['total_records']])
table.add_row(["Nama Atribut", f"{', '.join(info['attributes'])}"])
table.add_row(["Jumlah Attribute Objek", len(info['object_attributes'])])
table.add_row(["Jumlah Attribute Angka", len(info['number_attributes'])])
table.add_row(["Jumlah Attribute Null",({ len(info['null_attributes'])})])

print(table)
```

#### Ketentuan Analisis Dataset-3: Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka


```python
second_table = PrettyTable(["Atribut", "Data", "Hasil"])
second_table.align = "c"

attributes = list(df.select_dtypes(['int64', 'float64']).columns)


for attribute in attributes:
  df_atr = df[attribute]
  second_table.add_row([attribute, "Rata-rata", df_atr.mean() * 100])
  second_table.add_row(["", "Standar Deviasi", df_atr.std()])
  second_table.add_row(["", "Nilai Minimum", df_atr.min()])
  second_table.add_row(["", "Nilai Maksimum", df_atr.max()])
  second_table.add_row(["", "Q1", df_atr.quantile(.25)])
  second_table.add_row(["", "Q2", df_atr.quantile(.50)])
  second_table.add_row(["", "Q3", df_atr.quantile(.75)])
  second_table.add_row(["", "", ""])
  second_table.add_row(["", "", ""])

print(second_table)
```
